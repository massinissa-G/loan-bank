# Nettoyer la base donnée (Data Cleaning)

```python
#rensigner les valeurs manquantes et crée deux DataFrame (categorie,numérique)
cat_data=[]
num_data=[]
for i,c in enumerate(df.dtypes):
  if c==object:
    cat_data.append(df.iloc[:,i])
  else:
    num_data.append(df.iloc[:,i])
cat_data
```

```python
# Transposer nos DataFrame 
cat_data=pd.DataFrame(cat_data).transpose()
num_data=pd.DataFrame(num_data).transpose()
```

```python
cat_data
```

```python
num_data
```

```python
#pour les variables catégoriques on va remplacez les valeurs manquantes par les valeurs qui repete plus 
cat_data=cat_data.apply(lambda x:x.fillna(x.value_counts().index[0]))
cat_data.isnull().sum().any()
```

```python
#pour les variables numérique on va remplacez les valeurs manquantes par la valeur précedents de la meme colonne 
num_data.fillna(method='bfill',inplace=True)
num_data.isnull().sum().any() 
```

# Préparation du modele 

```python
# Diviser la base de données en base de donnée test et d'entrainement 
sss=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train,test in sss.split(X,y):
  X_train,X_test=X.iloc[train],X.iloc[test]
  y_train,y_test=y.iloc[train],y.iloc[test]

print('X_train taille: ', X_train.shape)
print('X_test taille: ', X_test.shape)
print('y_train taille: ', y_train.shape)
print('y_test taille: ', y_test.shape)
```

```python
# On va appliquer algorithme Logistic Regression 
models={'LogisticRegression':LogisticRegression(random_state=42)}

# La fonction de precision 
def accu(y_true,y_pred,retu=False):
  acc= accuracy_score(y_true,y_pred)
  if retu:
    return acc
  else:
    print(f'la precision de model est: {acc}')

#c'est la fonction d'application de model 
def train_test_eval(models,X_train,y_train,X_test,y_test):
  for name,model in models.items():
    print(name,':')
    model.fit(X_train,y_train)
    accu(y_test,model.predict(X_test))
    print('-'*30)

train_test_eval(models,X_train,y_train,X_test,y_test)
```

```python
# Le choix de ces 3 colonne parce que donne la meme performant que toute les colonne 
X_2=X[['Credit_History','Married','CoapplicantIncome']]

# Diviser la base de données en base de donnée test et d'entrainement 
sss=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train,test in sss.split(X_2,y):
  X_train,X_test=X_2.iloc[train],X_2.iloc[test]
  y_train,y_test=y.iloc[train],y.iloc[test]

print('X_train taille: ', X_train.shape)
print('X_test taille: ', X_test.shape)
print('y_train taille: ', y_train.shape)
print('y_test taille: ', y_test.shape)
```

```python
train_test_eval(models,X_train,y_train,X_test,y_test)
```
